```python
# <표2> 비선형성 검정
from statsmodels.tsa.stattools import bds
import numpy as np

def run_bds(series, max_dim=6):
    s = np.log(series.astype(float)).diff().dropna()
    s = s - s.mean()
    stat, pval = bds(s, max_dim=max_dim)
    return stat, pval


results_stat, results_pval = {}, {}

TARGET_COLS = df.columns.to_list()

for col in TARGET_COLS:
    stat, pval = run_bds(df[col])
    results_stat[col] = stat
    results_pval[col] = pval


index = [f"m={i}" for i in range(2, 2 + len(stat))]
bds_stat_df = pd.DataFrame(results_stat, index=index)
bds_pval_df = pd.DataFrame(results_pval, index=index)

print("\n<표 2> 비선형성 검정")
print("\n=== BDS 통계량 ===")
print(bds_stat_df.round(3))
print("\n=== BDS p-value ===")
print(bds_pval_df.applymap(lambda x: f"{x:.3e}"))
```

* 귀무가설 -> 선형변환으로 모든 의존성이 제거 가능
* 결과 해석 -> p < 0.05 -> 어떤 형태로든 비선형 상호의존 존재.

* p-value가 0.05보다 훨씬 작음 ⇒ 귀무가설(H0: 시계열이 i.i.d, 즉 “선형 + 백색잡음이면 충분”)을 기각

```python
# <표3> 단위근 검정

from statsmodels.tsa.stattools import adfuller, kpss

def adf(series):
    stat, p, *_ = adfuller(series, regression='c', autolag='AIC')
    return stat, p

TARGET_COLS = df.columns.to_list()

rows = {}
for col in TARGET_COLS :
    lvl_stat, lvl_p = adf(df[col].dropna())
    diff_stat, diff_p = adf(df[col].diff().dropna())
    rows[col] = [lvl_stat, lvl_p, diff_stat, diff_p]

tbl = pd.DataFrame(rows,
                   index=pd.MultiIndex.from_product(
                       [['Level','1st Difference'],
                        ['ADF Test Statistics','p-Value']]
                   )).T.round(4)

print("\n<표 3> 단위근 검정")
print(tbl.to_string(justify='center'))
```

```python
# ──────────────────────────────────────────────────────────────
# 0. IMPORTS
# ──────────────────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde, chi2
from tqdm import tqdm

# ──────────────────────────────────────────────────────────────
# 1. 핵심 함수 ─ KDE-기반 Transfer-Entropy (lag = 1)
# ──────────────────────────────────────────────────────────────
def te_kde(x: np.ndarray,
           y: np.ndarray,
           bw: str | float = 'scott') -> float:
    """
    Transfer Entropy  Y → X   (KDE, lag = 1)
    """
    x, y = map(lambda v: np.asarray(v, float), (x, y))
    mask = ~np.isnan(x) & ~np.isnan(y)
    x, y = x[mask], y[mask]
    if len(x) < 3 or np.allclose(x, x[0]) or np.allclose(y, y[0]):
        return 0.0

    x_t1, x_t, y_t = x[1:], x[:-1], y[:-1]     # lag = 1
    xyz = np.vstack([x_t1, x_t, y_t])          # (3, N)
    xz  = xyz[:2]                              # (2, N)

    pdf_xyz = gaussian_kde(xyz, bw_method=bw)
    pdf_xz  = gaussian_kde(xz , bw_method=bw)
    return float(np.mean(np.log((pdf_xyz(xyz)+1e-12)/(pdf_xz(xz)+1e-12))))

# ──────────────────────────────────────────────────────────────
# 2. 슬라이딩-윈도우 TE 시계열
# ──────────────────────────────────────────────────────────────
def sliding_te(df: pd.DataFrame,
               target: str,
               source: str,
               win: int = 36,
               step: int = 1,
               bw: str | float = 'scott',
               diff: int = 0) -> pd.Series:
    """
    return  TE(source → target)  시계열 (lag = 1, KDE)
    diff : 0 레벨, 1 1차 차분, 2 2차 차분
    """
    data = df[[target, source]].copy()
    if diff:
        data = data.diff(diff).dropna()

    x, y = data[target].values, data[source].values
    idx, vals = [], []

    for s in tqdm(range(0, len(data) - win - 1, step),
                  desc=f'TE({source}->{target})', leave=False):
        e = s + win
        vals.append(te_kde(x[s:e+1], y[s:e+1], bw))
        idx.append(data.index[s + win // 2])

    return pd.Series(vals, index=idx,
                     name=f'TE({source}->{target})')

# ──────────────────────────────────────────────────────────────
# Plotting
# ──────────────────────────────────────────────────────────────

# ──────────────────────────────────────────────────────────────
# 1. Z-score 표준화
# ──────────────────────────────────────────────────────────────
def zscore(s: pd.Series) -> pd.Series:
    """각 Series를 (값-평균)/표준편차 로 변환"""
    return (s - s.mean()) / s.std(ddof=0)

# ──────────────────────────────────────────────────────────────
# 2. 표준화 TE 시계열
# ──────────────────────────────────────────────────────────────
def compare_te(df: pd.DataFrame,
               extra_pairs: list[tuple[str, str]],
               win: int = 36,
               step: int = 1,
               bw: str | float = 'scott',
               diff: int = 0,
               colors=None,
               linestyles=None,
               *,
               standardize: bool = False):          # ← 추가
    """
    Transfer-Entropy 시계열 비교 플롯

    Parameters
    ----------
    standardize : bool, default False
        True → 각 TE 시계열을 Z-score 로 표준화해 그립니다.
    """
    base_src = '10년만기 국고채금리[미국]'
    base_tgt = '한국은행 기준금리'

    colors = colors or ['tab:blue', 'tab:orange', 'tab:green',
                        'tab:red', 'tab:purple', 'tab:brown']
    linestyles = linestyles or ['--', '-', '-.', ':',
                                (0, (5, 2)), (0, (3, 1, 1, 1))]

    fig, ax = plt.subplots(figsize=(10, 4))

    # ── 베이스 라인 ──────────────────────────────────────────
    base_series = sliding_te(df, base_tgt, base_src,
                             win, step, bw, diff)
    if standardize:
        base_series = zscore(base_series)

    plot_te(base_series, ax=ax,
            label=f'TE({base_src}→{base_tgt})',
            color='black', linestyle='--', lw=2)

    # ── 추가 비교쌍 ───────────────────────────────────────────
    for i, (src, tgt) in enumerate(extra_pairs):
        s = sliding_te(df, tgt, src, win, step, bw, diff)
        if standardize:
            s = zscore(s)

        plot_te(s, ax=ax,
                label=f'TE({src}→{tgt})',
                color=colors[i % len(colors)],
                linestyle=linestyles[i % len(linestyles)])

    ax.set_title('Transfer Entropy Comparison' +
                 (' (표준화)' if standardize else ''))
    ax.legend()
    plt.tight_layout()
    return ax
```

```python
# ────────────────────────────────────────────────────────────────
# 0. 필수 import
# ────────────────────────────────────────────────────────────────
import numpy as np
import pandas as pd
from tqdm import tqdm
from scipy.stats import gaussian_kde, chi2

# ────────────────────────────────────────────────────────────────
# 1.  KDE 기반 Transfer-Entropy  (lag = 1)
# ────────────────────────────────────────────────────────────────
def te_kde(x: np.ndarray,
           y: np.ndarray,
           bw: str | float = 'scott') -> float:
    """
    Transfer Entropy  Y → X  (lag = 1, KDE)
    · NaN/상수/샘플 부족/공분산 singular → 0 반환
    """
    x, y = map(np.asarray, (x, y))
    mask = ~np.isnan(x) & ~np.isnan(y)
    x, y = x[mask], y[mask]
    if len(x) < 3 or np.allclose(x, x[0]) or np.allclose(y, y[0]):
        return 0.0

    x_t1, x_t, y_t = x[1:], x[:-1], y[:-1]
    data_xyz = np.vstack([x_t1, x_t, y_t])  # (3, N)
    data_xz  = data_xyz[:2]                 # (2, N)

    try:
        pdf_xyz = gaussian_kde(data_xyz, bw_method=bw)
        pdf_xz  = gaussian_kde(data_xz , bw_method=bw)
        return float(np.mean(np.log((pdf_xyz(data_xyz)+1e-12) /
                                    (pdf_xz (data_xz )+1e-12))))
    except np.linalg.LinAlgError:           # singular cov
        return 0.0

# ────────────────────────────────────────────────────────────────
# 2.  윈도우별 TE → LLR 표 생성
# ────────────────────────────────────────────────────────────────
def llr_kde(df: pd.DataFrame,
            target: str,
            sources: list[str],
            win: int          = 36,
            step: int         = 1,
            bw: str | float   = 'scott',
            surro_iter: int   = 100,
            diff_order: int   = 0,
            shuffle: bool     = True) -> pd.DataFrame:
    """
    · df         : datetime index DataFrame
    · target     : X 변수
    · sources    : Y 변수 리스트  (target 은 포함 X)
    · diff_order : 0 레벨 / 1 1차 / 2 2차
    · shuffle    : True → Y를 무작위 셔플, False → 순환 shift
    반환 : MultiIndex(변수, [LLR,P*]) 표
    """
    data = df[[target]+sources].copy()
    if diff_order:
        data = data.diff(diff_order)
    data = data.dropna()

    # 상수열 자동 제거
    sources = [s for s in sources if data[s].nunique() > 1]

    rows, dates = [], []
    for st in tqdm(range(0, len(data)-win, step), desc='windows', leave=False):
        ed   = st + win
        x_w  = data[target].iloc[st:ed+1].values
        row  = {}
        for src in sources:
            y_w   = data[src].iloc[st:ed+1].values
            te_t  = te_kde(x_w, y_w, bw)

            sur   = [te_kde(x_w,
                            np.random.permutation(y_w) if shuffle
                            else np.roll(y_w, k+1),
                            bw)
                     for k in range(surro_iter)]
            te_s  = np.mean(sur)

            N     = win                        # 샘플 수
            llr   = max(0, 2*N*(te_t - te_s))  # 음수 → 0
            p_val = 1 - chi2.cdf(llr, 1)

            row[(src, 'LLR')] = round(llr, 1)
            row[(src, 'P*' )] = f"{p_val:.2f}"
        rows.append(row)
        dates.append(data.index[st + win//2])

    return pd.DataFrame(rows, index=dates,
                        columns=pd.MultiIndex.from_product(
                            [sources, ['LLR', 'P*']])
                       )
```

```python
# ────────────────────────────────────────────────────────────────
# 3.  사용 예시
# ────────────────────────────────────────────────────────────────

# ① 데이터 읽기

# ② 분석 변수
target  = '한국은행 기준금리'
sources = [c for c in df.columns if c != target]

# ③ LLR 표 생성
llr_tbl = llr_kde(df, target, sources,
                  win=36, step=1,
                  bw='scott', surro_iter=100,
                  diff_order=0, shuffle=True)

print(llr_tbl.head(10))      # 미리보기
llr_tbl.to_csv('LLR_table.csv', encoding='utf-8-sig')
```

```python
llr = pd.read_csv('/Users/sanghyun/Desktop/paper/LLR_table.csv', header=[0, 1], index_col=0)

# p-value 열만 선택 → float 변환
p_cols = [c for c in llr.columns if c[1] == 'P*']
pvals  = llr[p_cols].astype(float)

# 각 변수별 p < 0.05 비중
crit         = 0.01
sig_ratio    = (pvals < crit).sum() / len(pvals)
sig_ratio_df = sig_ratio.sort_values(ascending=False) \
                        .to_frame('sig_ratio')

print(sig_ratio_df)
```

```python
sig_var = (sig_ratio_df
           .loc[np.isclose(sig_ratio_df['sig_ratio'], 1) |  # 1.0 포함
                (sig_ratio_df['sig_ratio'] >= 0.89)]         # 0.8 이상
           .index.get_level_values(0)                       # ← 레벨 0
           .unique()                                        # 중복 제거
           .tolist())
sig_var
```

```python
from arch import arch_model

def make_lags(series: pd.Series, max_lag: int) -> pd.DataFrame:
    """
    단일 시계열을 1‥max_lag 까지 시차 변수로 변환
    """
    return pd.concat({f'{series.name}(-{i})': series.shift(i)
                      for i in range(1, max_lag + 1)}, axis=1)

def preprocess(df_raw: pd.DataFrame) -> pd.DataFrame:
    """논문 방식 전처리:
       - 금리: 수준, 그 외: 로그변환 후 1차 차분"""
    df = df_raw.copy()
    price_like = df.columns
    for col in price_like:
        df[col] = np.log(df[col]).diff()
    df = df.dropna()
    return df


def fit_bok_equation1(df: pd.DataFrame,
                      ar_lag: int = 1,
                      exo_lag: int = 0,
                      garch_p: int = 1,
                      garch_o: int = 1,
                      garch_q: int = 1):
    """
    BOK Equation 1 + EGARCH(p,o,q) 추정
    """
    y = df['한국은행 기준금리']

    # --- AR(1) 부분 --------------------
    ar_terms = make_lags(y, ar_lag)

    # --- exogenous (동일시점 + 시차 exo_lag) ------------
    x_cols_now = sig_var

    x_now = df[x_cols_now]
    if exo_lag > 0:
        lag_frames = [make_lags(df[col], exo_lag) for col in x_cols_now]
        x_all = pd.concat([x_now] + lag_frames, axis=1)
    else:
        x_all = x_now


    design = pd.concat([y, ar_terms, x_all], axis=1).dropna()
    y_aligned = design['한국은행 기준금리']
    X_aligned = design.drop(columns='한국은행 기준금리')

    # --- EGARCH ----------------------------
    model = arch_model(y_aligned,
                       x=X_aligned,
                       mean='ARX', lags=0,      # AR은 직접 만들어 넣었으므로 0
                       vol='EGARCH',
                       p=garch_p, o=garch_o, q=garch_q,
                       dist='normal')

    res = model.fit(show_warning=False, disp='off')
    return res
```

```python
df = preprocess(df_BOK_1)

#3) EGARCH 추정 --------------------------------------------------------
result = fit_bok_equation1(df,
                           ar_lag=1,   # DBOK(-1) : 1 고정
                           exo_lag=3,  # 시
                           garch_p=1, garch_o=1, garch_q=1)

# 4) 결과 요약 -----------------------------------------------------------
print(result.summary())
```
